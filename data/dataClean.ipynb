```python
df = df.drop(columns=['link', 'ram_type_tokenized', 'gpu_processor tokenized', 'Refresh Rate'])
df.head()
```

```python
category_mapping = {
    1: 'gaming',
    2: 'thin&light',
    3: '2in1',
    4: 'notebook',
    5: 'laptop',
    6: '2in1gaming',
    7: 'business',
    8: 'chromebook',
    9: 'creator'
}

df['Type'] = df['Type'].replace(category_mapping)
df.head()
```

```python
category_mapping = {
    1: 'intel',
    2: 'amd',
    3: 'qualcomm',
    4: 'apple',
    5: 'mediatek',
}

df['Processor Brand'] = df['Processor Brand'].replace(category_mapping)
df.head()
```

```python
category_mapping = {
    0: 'no',
    1: 'yes',
}

df['SSD'] = df['SSD'].replace(category_mapping)
df['Expandable Memory'] = df['Expandable Memory'].replace(category_mapping)
df['Touchscreen'] = df['Touchscreen'].replace(category_mapping)
df.head()
```

```python
category_mapping = {
    1: 'windows',
    2: 'chromeos',
    3: 'dos',
    4: 'mac',
    5: 'ubuntu',
}

df['Operating System'] = df['Operating System'].replace(category_mapping)
df.head()
```

```python

category_mapping = {
    1: 'asus',
    2: 'hp',
    3: 'lenovo',
    4: 'dell',
    5: 'msi',
    6: 'realme',
    7: 'avita',
    8: 'acer',
    9: 'samsung',
    10: 'infinix',
    11: 'lg',
    12: 'apple',
    13: 'nokia',
    14: 'redmibook',
    15: 'mi',
    16: 'vaio'
}

df['company'] = df['company'].replace(category_mapping)
df.head()
```
