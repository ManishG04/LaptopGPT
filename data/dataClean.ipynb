## Feature Engineering

```python

cpu_scaler = MinMaxScaler(feature_range=(0, 1))
df['Normalized_CPU_Ranking'] = 1 - cpu_scaler.fit_transform(df[['CPU_ranking']])

gpu_scaler = MinMaxScaler(feature_range=(0, 1))
df['Normalized_GPU_Benchmark'] = gpu_scaler.fit_transform(df[['gpu_benchmark']])

max_weight = df["Weight (in kg)"].max()
max_battery = df["battery_backup"].max()
max_screen_size = df["Screen Size (in inch)"].max()

w1, w2, w3 = 0.4, 0.3, 0.3
df["Portability"] = (
    w1 * (1 - df["Weight (in kg)"] / max_weight) +           
    w2 * (df["battery_backup"] / max_battery) +             
    w3 * (1 - df["Screen Size (in inch)"] / max_screen_size)
)


min_portability = df["Portability"].min()
max_portability = df["Portability"].max()
df["Portability"] = 1 + 99*(df["Portability"] - min_portability) / (max_portability - min_portability)


weights = {
    'CPU': 0.3,
    'GPU': 0.3,
    'RAM': 0.2,
    'Dedicated_Graphics': 0.2
}
df['Performance_Score'] = (
    weights['CPU'] * df['Normalized_CPU_Ranking'] +
    weights['GPU'] * df['Normalized_GPU_Benchmark'] +
    weights['RAM'] * df['RAM (in GB)'] / df['RAM (in GB)'].max() +  # Normalize RAM inline
    weights['Dedicated_Graphics'] * df['Dedicated Graphic Memory Capacity'] / df['Dedicated Graphic Memory Capacity'].max()
)

df["Performance_Score"] = 1 + 99 * (df["Performance_Score"])


df['Value_Score'] = df['Performance_Score'] / df['Price (in Indian Rupees)']
min_value_score = df["Value_Score"].min()
max_value_score = df["Value_Score"].max()

df["Value_Score"] = 1 + 99*(df["Value_Score"] - min_value_score) / (max_value_score - min_value_score)



# Display the resulting dataframe with new columns
df
```
